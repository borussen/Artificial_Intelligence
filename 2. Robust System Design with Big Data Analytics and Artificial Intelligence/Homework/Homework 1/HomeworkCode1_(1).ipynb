## 데이터 불러오기

ExampleData 불러오기

```python
ExampleData = pd.read_csv('./ExampleData', sep=',', header=None)
ExampleData
```

0.01초부터(이상, 0.01초 포함) 0.09초까지(미만, 0.09초 미포함) 해당하는 데이터 추출

```python
StartPoint = np.where(ExampleData.iloc[:,0].values == 0.01)[0][0]
EndPoint   = np.where(ExampleData.iloc[:,0].values == 0.09)[0][0]

StartPoint, EndPoint
```

```python
NewData = ExampleData.iloc[StartPoint:EndPoint, :]
NewData
```

## 데이터 핸들링 (Handling)

시간열(time column) 제거한 센서 데이터 추출

```python
SensorDataOnly = NewData.iloc[:,1:]
SensorDataOnly
```

데이터를 파일(.csv)로 저장

```python
path = './ST136_HW1_1'
SensorDataOnly.to_csv(path , sep=',' , header=None , index=None)
```
